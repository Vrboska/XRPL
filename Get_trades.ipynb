```python
from xrpl.models.requests import AccountTx
from xrpl.models.transactions import OfferCreate
from xrpl.asyncio.clients import AsyncJsonRpcClient

# Connect to the XRP Ledger (testnet)
client = AsyncJsonRpcClient("https://xrplcluster.com")  # Mainnet
#client = AsyncJsonRpcClient("https://s.altnet.rippletest.net:51234")

# Define the currencies for the trading pair
XRP = {"currency": "XRP"}  # XRP is the native currency
USD = {"currency": "USD", "issuer": "rvYAfWj5gh67oV6fW32ZzP3Aw4Eubs59B"}  # Replace with the actual USD issuer address


async def get_trades():
# Fetch recent transactions from the ledger
    account_tx_request = AccountTx(
        account="rvYAfWj5gh67oV6fW32ZzP3Aw4Eubs59B",  # Leave empty to fetch all transactions
        ledger_index_min=-1,
        ledger_index_max=-1,
        limit=1000, # Adjust the limit as needed,
        binary=False
    )

    response = await client.request(account_tx_request)
    #print(response)
    # Process transactions and filter trades
    transactions = response.result["transactions"]
    #print(transactions)
    trades = [
        tx for tx in transactions
        if isinstance(tx, dict) and "tx_json" in tx and tx["tx_json"].get("TransactionType") in ["OfferCreate", "OfferCancel", "OfferExecute"]
    ]

    # Print trade details
    for trade in trades:
        #print(trade)
        tx_data = trade["tx_json"]
        print(f"Transaction Type: {tx_data.get('TransactionType')}")
        print(f"Date: {tx_data.get('date')}")
        print(f"Account: {tx_data.get('Account')}")
        print(f"Offer Details: {tx_data.get('TakerGets', '')} -> {tx_data.get('TakerPays', '')}")
        print(f"Cancelling OfferSequence: {tx_data.get('OfferSequence')}")
        print(f"Sequence: {tx_data.get('Sequence')}")
        print("=" * 40)
        
await get_trades()
```
